This class maps the 'ant_bait.csv' structure to the python class named 'ant_bait'. This class has one method for printing and one constructor. The constructor receives all the attributes as parameter.

````python
class ant_bait:
    def __init__(self, row):
        self.month =row[0]
        self.year  =row[1]
        self.plot  =row[2]
        self.stake =row[3]
        self.species =row[4]
        self.abundance =row[5]
    def output_print(self):
        print(str(self.species) +" - " +str(self.month) +" , " +str(self.year) +" : " +str(self.abundance))
        
````

This class maps the 'ant_species.csv' structure to the python class named 'ant_species'. This class has one method for printing and one constructor. The constructor receives all the attributes as parameter.

````python
class ant_species:
    def __init__(self, row):
        self.speciescode =row[0]
        self.genus       =row[1]
        self.altgenus    =row[2]
        self.species     =row[3]
        self.altspecies  =row[4]
        self.tribe       =row[5]
        self.subfamily   =row[6]
        self.IDissues    =row[7]
    def output_print(self):
        print(str(self.genus) +" " +str(self.species))
    
````

This function creates object of each row of the 'ant_bait.csv' file. It takes the filename as an argument. It appends each of the object (created from each row) to a list called 'bait_list'.

````python
def ant_bait_function(filename):
    bait_list =[]
    try:
        with open(filename) as csvfile:
            readCSV = csv.reader(csvfile, delimiter=',')
            next(csvfile)
            for row in readCSV:
                a =ant_bait(row)
                bait_list.append(a)
        return bait_list
    except:
        print('Error in Filename')
        return []
````

This function creates object of each row of the 'ant_species.csv' file. It takes the filename as an argument. It appends each of the object (created from each row) to a list called 'species_list'.

````python
def ant_species_function(filename):
    species_list =[]
    try:
        with open(filename) as csvfile:
            readCSV = csv.reader(csvfile, delimiter=',')
            next(csvfile)
            for row in readCSV:
                a =ant_species(row)
                species_list.append(a)
        return species_list
    except:
        print('Error in Filename')
        return []
    
````

This function replaces the attribute 'species' of ant_bait with 'speciescode' of 'ant_species' 

````python
def attribute_replacement(bait_list, species_list):
    speciescodes =[]
    for elements in species_list:
        speciescodes.append(elements.speciescode)
    for elements in bait_list:
        elements.species = species_list[speciescodes.index(elements.species)]
````

This chunk of code is used to check all the above functions and classes. Here I checked only the 'attribute_replacement'. As it uses other functions and objects, only checking this one validates the whole program. 

````python
bait_list = ant_bait_function('ant_bait.csv')  
species_list = ant_species_function('ant_species.csv')
attribute_replacement(bait_list, species_list)
for elements in bait_list:
    elements.species.output_print()
````
